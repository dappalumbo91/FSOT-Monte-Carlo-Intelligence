```python
rng = np.random.default_rng(42)
r = rng.normal(0.0004, 0.012, size=400)
close = 100 * np.cumprod(1 + r)
df = pd.DataFrame({
    'time': pd.date_range('2020-01-01', periods=400, freq='B'),
    'open': close, 'high': close*1.01, 'low': close*0.99,
    'close': close, 'volume': rng.uniform(1e6, 2e6, size=400),
})
intel = run_intelligence(df, symbol='SYN', horizon=13, n_paths=128, include_bhs=False)
print('primary', intel['primary_signal'], intel['primary_confidence'])
print('mc', intel['monte_carlo']['signal'], intel['monte_carlo']['confidence'])
print('authority_ok', intel['authority_gate']['ok'])
```
